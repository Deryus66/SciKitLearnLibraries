```python
df = pd.read_csv('covid-death-data.csv')
df.head()
```

```python
newdf = df.fillna(0)
newdf.head()
newdf['covid_deaths_per_500k'] = newdf['covid_deaths_per_500k'].str.replace(',','').astype(float)
newdf.loc[741]
newdf['covid_deaths_per_500k'] = newdf['covid_deaths_per_500k'].astype(float)
newdf['covid_inpatient_bed_utilization'] = newdf['covid_inpatient_bed_utilization'].astype(float)
```

```python
x = newdf['covid_inpatient_bed_utilization']
y = newdf['covid_deaths_per_500k']

plt.scatter(x, y)
plt.ylim(0, 1000)
plt.show()
```

```python
km = KMeans(n_clusters=6)
predictedY = km.fit_predict(newdf[['covid_inpatient_bed_utilization','covid_deaths_per_500k']])
predictedY
```

```python
newdf['cluster'] = predictedY
newdf.head()
```

```python
df1 = newdf[newdf.cluster==0]
df2 = newdf[newdf.cluster==1]
df3 = newdf[newdf.cluster==2]

df4 = newdf[newdf.cluster==3]
df5 = newdf[newdf.cluster==4]
df6 = newdf[newdf.cluster==5]

plt.scatter(df1['covid_inpatient_bed_utilization'], df1['covid_deaths_per_500k'], color = 'green')
plt.scatter(df2['covid_inpatient_bed_utilization'], df2['covid_deaths_per_500k'], color = 'red')
plt.scatter(df3['covid_inpatient_bed_utilization'], df3['covid_deaths_per_500k'], color = 'black')

plt.scatter(df4['covid_inpatient_bed_utilization'], df4['covid_deaths_per_500k'], color = 'blue')
plt.scatter(df5['covid_inpatient_bed_utilization'], df5['covid_deaths_per_500k'], color = 'orange')
plt.scatter(df6['covid_inpatient_bed_utilization'], df6['covid_deaths_per_500k'], color = 'yellow')
plt.ylim(0, 1000)
```
